```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import *
import ipywidgets as ipw
```

```python
t=1
length=25.15
pi=np.pi
x=np.linspace(0,length,1001)
```

```python
def standing_wave(freq):
    global df
    incident_wave  = np.sin(x+freq*t)
    reflected_wave = np.sin(x-freq*t)    
    result_wave=incident_wave + reflected_wave 
    s_x=pd.Series(x)
    s_inci=pd.Series( incident_wave)
    s_refl=pd.Series( reflected_wave)
    s_res = pd.Series( result_wave)
    st=pd.concat([s_x,s_inci,s_refl,s_res],axis=1)
    cols=['x_position','incident_wave','reflected_wave','resulting_wave']
    st.columns=cols
    f=plt.figure(figsize=(16,9))
    plt.style.use('fivethirtyeight')
    plt.title('res_wave =' + str(round(np.max(s_res),2)) + 
              ' Sin (' + str(round(2*pi*freq,2)) + ' t +phi' +')',size =25)
    my_label="$resulting$_$wave$_$peak$ = {} "
    plt.plot(x, incident_wave,color='blue',lw=6,label="$incident$_$wave$")
    plt.plot(x, reflected_wave,color='green',lw=5,label="$reflected$_$wave$")
    plt.plot(x, result_wave,color='chocolate',lw=8,
             label=my_label.format(round(np.max(result_wave),4)))
    plt.legend(loc='upper right')
    plt.xlabel('$x$_$position[cm]$',size=25)
    plt.ylabel('$amplitude[cm]$',size=25)
    plt.ylim(-2.1,2.1)
    plt.tick_params(which='major', length=10, width=0.9, direction='out')
    plt.tick_params(which='minor', length=10, width=0.9, direction='out')
    plt.grid(linestyle='--')
    plt.show()
    df=pd.DataFrame(st)
    #f.savefig('standing_code_11.'+str(freq)+'.png')
    #df.to_csv('standing_waves.csv')
    #return st.head(2),st.tail(2)
```

```python
freq=FloatSlider(description='$frequency$',min=0.,max=30.1,step=0.1,value=0)
interact(standing_wave,freq=freq);
```
